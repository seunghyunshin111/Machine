```
import tensorflow as tf
from sklearn.decomposition import PCA
```

```
(x_train, _y), (x_test, _) = tf.keras.datasets.boston_housing.load_data()
x_train.shape
```

```
import numpy as np
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer().data
target = load_breast_cancer().target
data = data[:, [4,5]]
```

```
import matplotlib.pyplot as plt
plt.scatter(data[:, 0], data[:, 1], c=target, alpha=0.6)
```

```
pca = PCA(n_components=1)
transformed_data = pca.fit_transform(data)
transformed_data.shape
```

```
plt.scatter(transformed_data, np.zeros(len(transformed_data)), c=target, alpha=0.6)
```

```
# tensorflow AutoEncoder(Linear)
data.shape
```

```
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(2)                         
])

# x = tf.keras.layers.Dense(1)(data)
# output = tf.keras.layers.Dense(2)(data)
# tf.keras.Model(data, output)

model.compile(loss='mse')
model.fit(data, data, epochs=3000)
```

```
plt.subplot(1,2,1)
plt.scatter(model.layers[0](data), np.zeros(len(data)), c=target, alpha=0.6)
plt.subplot(1,2,2)
plt.scatter(transformed_data, np.zeros(len(transformed_data)), c=target, alpha=0.6)
```

```
plt.scatter(model.predict(data)[:, 0], model.predict(data)[:, 1], c=target)
```
